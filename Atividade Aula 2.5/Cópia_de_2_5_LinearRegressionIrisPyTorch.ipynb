# Regressão Linear com PyTorch

## Objetivo

Este exemplo utiliza regressão linear para estimar o comprimento das sépalas da íris a partir do seu comprimento das pétalas.
Utiliza-se
- o cálculo do gradiente automático do PyTorch, porém
- a função de perda e
- atualização dos parâmetros é feita de forma explícita.
- A rede é criada com uma camada nn.Linear()

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinear_Otimizacao.png)

## Importação dos pacotes

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
from sklearn.datasets import load_iris

import torch
from torch import nn, optim
from torch.autograd import Variable

torch.manual_seed(1234)
```

## Leitura dos dados

```python
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

x_train = data[:,0:1].astype(np.float32)
y_train = data[:,1:2].astype(np.float32)

n_samples = x_train.shape[0]
print('x_train.shape:',x_train.shape, x_train.dtype)
print('y_train.shape:',y_train.shape, y_train.dtype)

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

### Normalização dos dados

```python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()
```

```python
print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

```python
x_train_bias = np.hstack([np.ones(shape=(n_samples,1)), x_train])
```

```python
x_train_bias
```

```python
print('x_train_bias.shape:',x_train_bias.shape, x_train_bias.dtype)
```

```python
x_train_bias = torch.FloatTensor(x_train_bias)
y_train      = torch.FloatTensor(y_train)
print(y_train)
```

## Model

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinearNeuronio.png)

```python
#Inicialização usando toch.nn.Linear

model = torch.nn.Linear(in_features=2, out_features=1, bias=False)
```

### Verificando a inicialização dos parâmetros

```python
model.weight.data = torch.zeros(1,2)
torch.nn.init.uniform(model.weight.data, -0.1, 0.1)
model.weight
```

### Testando o predict da rede

```python
xx = torch.ones((5,2))
xx
```

```python
model(Variable(xx))
```

## Treinamento

```python
num_epochs = 100
learning_rate = 0.5
for epoch in range(num_epochs):
    #inputs = Variable(x_train_bias)
    #target = Variable(y_train)
    inputs = x_train_bias
    target = y_train

    # forward - predict
    y_pred = model(inputs)

    #loss cálculo da função de perda
    loss = torch.mean((y_pred - target)**2)

    # backward propagation - cálculo gradient
    loss.backward()

    # gradiente descendente
    model.weight.data = model.weight.data - learning_rate * model.weight.grad.data

    # zeragem parâmetros próximo cálculo do gradiente
    model.weight.grad.data.zero_()

    # verbose
    if (epoch+1) % 1 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch+1, num_epochs, loss.data))
```

## Avaliação

```python
y_pred = model(Variable(x_train_bias))
plt.plot(x_train, y_train.numpy(), 'ro', label='Original data')
plt.plot(x_train, y_pred.data.numpy(), 'kx-', label='Fitting Line')
plt.show()
```

### Parâmetros ótimos, solução analítica

```python
x_bias = x_train_bias
y = y_train
x_bias_t = torch.t(x_bias)
w_opt = (torch.inverse(x_bias_t.mm(x_bias)).mm(x_bias_t)).mm(y)
print(torch.t(w_opt))
```

# Exercícios

- 1) Por que no final da varredura do laço de treinamento, a perda é impressa utilizando
   `loss.data`? Por que precisa do `data`?
   

# Porque Loss é um tensor que também guarda o grafo de gradiente. Com o '.data' acessamos apenas o valor numérico da perda.

- 2) Calcule e imprima o valor da perda (MSE) fazendo o predict da rede com os pesos já
   treinados.

```python
# Faz o predict com os pesos treinados
y_pred = model(x_train_bias)

# Calcula a perda (Mean Squared Error)
loss = torch.mean((y_pred - y_train) ** 2)

# Imprime o valor da perda
print('MSE final:', loss.item())
```

- 3) Crie duas listas `w0_list` e `w1_list` para guardar a evolução dos valores dos
   parâmetros durante o laço de treinamento.
   Modifique o exemplo do gráfico tipo **scatter plot** para visualizar a variação
   dos parâmetros do modelo duas dimensões.
   Marque neste gráfico o valor ótimo do parâmetro dado pela solução
   analítica.

```python
# listas para rastrear a trajetória dos pesos
w0_list, w1_list = [], []

# salvar estado inicial ANTES do treino
with torch.no_grad():
    w0_list.append(model.weight[0, 0].item())  # w0 (bias)
    w1_list.append(model.weight[0, 1].item())  # w1 (coeficiente)

num_epochs = 100
learning_rate = 0.5

for epoch in range(num_epochs):
    inputs = x_train_bias  # [1, x]
    target = y_train

    # forward
    y_pred = model(inputs)

    # MSE
    loss = torch.mean((y_pred - target) ** 2)

    # backward
    loss.backward()

    # atualização (gradient descent "na mão")
    with torch.no_grad():
        model.weight -= learning_rate * model.weight.grad

    # zerar gradientes
    model.weight.grad.zero_()

    # registrar APÓS a atualização desta época
    with torch.no_grad():
        w0_list.append(model.weight[0, 0].item())
        w1_list.append(model.weight[0, 1].item())


# Renomeando parâmetros da solução ótima analítica
with torch.no_grad():
    w0_star = w_opt[0, 0].item()  # ótimo para o bias
    w1_star = w_opt[1, 0].item()  # ótimo para a feature


# índices das épocas desejadas
# lembrando que w0_list[0] é antes do treino começar
idxs = [1, 33, 66, 100]

# selecionar apenas os valores correspondentes
w0_plot = [w0_list[i] for i in idxs]
w1_plot = [w1_list[i] for i in idxs]

# criar figura
plt.figure(figsize=(6, 5))

# pontos das 4 épocas
plt.scatter(w0_plot, w1_plot, label="Épocas 1, 33, 66 e 100", color='blue')

# setas entre as épocas
w0_old, w1_old = None, None
for w0, w1 in zip(w0_plot, w1_plot):
    if w0_old is not None:
        plt.arrow(w0_old, w1_old, w0 - w0_old, w1 - w1_old,
                  head_length=0.01, head_width=0.01, shape='full',
                  length_includes_head=True, color='blue')
    w0_old, w1_old = w0, w1

# marcar o ponto ótimo (solução analítica)
plt.scatter([w0_star], [w1_star], marker='*', s=200, edgecolors='k',
            color='gold', label='Solução Analítica (ótimo)')

# rótulos e título
plt.xlabel("w0 (bias)")
plt.ylabel("w1")
plt.title("Evolução dos parâmetros (épocas 1, 33, 66 e 100)")
plt.legend()
plt.tight_layout()
plt.show()
```
